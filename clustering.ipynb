```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset from a CSV file
# Replace 'your_file.csv' with the path to your actual CSV file
df = pd.read_csv('subset.csv')

# Preview the first few rows of the dataset to understand its structure
print(df.head())

# Assuming the two columns to cluster are named 'Column1' and 'Column2'
# If your columns have different names, replace 'Column1' and 'Column2' with the correct names from your dataset
X = df[['Total_Count', 'Close_Delta']]

# Apply KMeans clustering (let's assume 2 clusters here)
kmeans = KMeans(n_clusters=2, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Plot the data points and the cluster centers
plt.figure(figsize=(8, 6))
plt.scatter(df['Total_Count'], df['Close_Delta'], c=df['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200)
plt.title('KMeans Clustering of Two Columns')
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.show()

# Print the DataFrame with cluster labels
print(df)
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset from a CSV file
# Replace 'your_file.csv' with the path to your actual CSV file
df = pd.read_csv('data/JNJ_merged.csv')
df['Insider_Trades'] = df['Insider_Trades'].fillna(0)
df['Close_Delta'] = df['Close_Delta'].fillna(0)

# Preview the first few rows of the dataset to understand its structure
# print(df.head())

# Define the columns to cluster
col1, col2 = 'Close_Delta', 'Insider_Trades'

# Ensure the column names are correct and match exactly as in your dataset
X = df[[col1, col2]]

print(X.info())

# Apply KMeans clustering (let's assume 2 clusters here)
kmeans = KMeans(n_clusters=2, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Plot the data points and the cluster centers
plt.figure(figsize=(8, 6))
plt.scatter(df[col1], df[col2], c=df['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200)
plt.title('KMeans Clustering of Two Columns')
plt.xlabel(col1)
plt.ylabel(col2)
plt.show()

# Print the DataFrame with cluster labels
print(df)
```
